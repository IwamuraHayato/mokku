```python
# DataFrameを作成
df = pd.DataFrame(contents)

# 'plain_text' だけを抽出する関数を定義

def extract_plain_text(cell):
    if isinstance(cell, dict):  # cellが辞書であるか確認
        rich_text_content = cell.get('rich_text', [])
        if isinstance(rich_text_content, list) and len(rich_text_content) > 0:
            # 'rich_text' がリストで、かつそのリストが空でないか確認
            return rich_text_content[0].get('plain_text', None)
    return None  # 該当がない場合は None を返す

# 各セルに 'plain_text' を適用
df_plain_text = df.applymap(extract_plain_text)

# 結果の表示
df_plain_text
```

```python
# csvへ出力
df_plain_text.to_csv('output.csv', index=False)
```
